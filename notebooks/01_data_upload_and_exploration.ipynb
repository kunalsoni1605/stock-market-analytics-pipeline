```python
# Load the data
df = spark.table("workspace.default.stock_prices_raw")
```

```python
display(df)

print(f" Data loaded successfully!")
print(f"📊 Total records: {df.count():,}")
```

```python
# Check the structure of data
print("=" * 70)
print("📋 DATA SCHEMA")
print("=" * 70)
df.printSchema()

print("\n" + "=" * 70)
print("📊 COLUMN NAMES")
print("=" * 70)
print(df.columns)
```

```python
# Data Summary Statistics
from pyspark.sql.functions import count, countDistinct, min, max

print("=" * 70)
print("📊 DATA SUMMARY")
print("=" * 70)

# Total records
total = df.count()
print(f"Total records: {total:,}")

# Unique stocks
stocks = df.select(countDistinct("Symbol")).collect()[0][0]
print(f"Number of stocks: {stocks}")

# Date range
dates = df.select(min("date"), max("date")).collect()[0]
print(f"Date range: {dates[0]} to {dates[1]}")

print("\n📈 Stocks tracked:")
df.select("Symbol").distinct().orderBy("Symbol").show()
```

```python
#Records Per Stock
print("=" * 70)
print("📊 RECORDS PER STOCK")
print("=" * 70)

records_per_stock = df.groupBy("Symbol") \
    .count() \
    .orderBy("count", ascending=False)

display(records_per_stock)
```

```python
#Latest Stock Prices
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col, desc

print("=" * 70)
print("💰 LATEST CLOSING PRICES")
print("=" * 70)

# Get the most recent price for each stock
window = Window.partitionBy("Symbol").orderBy(desc("date"))

latest_prices = df.withColumn("rank", row_number().over(window)) \
    .filter(col("rank") == 1) \
    .select("Symbol", "date", "close_price") \
    .orderBy("Symbol")

display(latest_prices)
```

```python
#Check for Missing Values
from pyspark.sql.functions import col, sum as spark_sum

print("=" * 70)
print("🔍 MISSING VALUES CHECK")
print("=" * 70)

# Count nulls in each column
missing_counts = df.select([
    spark_sum(col(c).isNull().cast("int")).alias(c) 
    for c in df.columns
])

display(missing_counts)
```
